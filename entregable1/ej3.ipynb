```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función
def funcion_objetivo(x):
    return 4*x**4 - x**3 - 4*x**2 + 1

# Generar valores de x
x = np.linspace(-1.2, 1.2, 400)  # Valores de x en el rango de interés

# Calcular los valores de y (f(x)) correspondientes
y = funcion_objetivo(x)

# Crear la gráfica
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Función $4x^4 - x^3 - 4x^2 + 1$')
plt.title('Gráfica de la función objetivo')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función
def funcion_objetivo3(x, a):
    return (x - a)**2 + 1

# Generar valores de x
x = np.linspace(-10, 10, 400)  # Valores de x en el rango de interés

# Valor de a (puedes ajustar este valor)
a = 2

# Calcular los valores de y (f(x)) correspondientes
y = funcion_objetivo3(x, a)

# Crear la gráfica
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'Función $(x - {a})^2 + 1$')
plt.title('Gráfica de la función objetivo')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Definir la función
def funcion_objetivo(x, a):
    return (x - a)**2 + 1

# Generar valores de x
x = np.linspace(-10, 10, 400)  # Valores de x en el rango de interés

# Valores de a
valores_a = [0, 2, -1]

# Crear la gráfica
plt.figure(figsize=(8, 6))

# Dibujar las curvas para diferentes valores de a
for a in valores_a:
    y = funcion_objetivo(x, a)
    plt.plot(x, y, label=f'$(x - ({a}))^2 + 1$')

plt.title('Gráfica de diferentes curvas $(x - a)^2 + 1$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```
