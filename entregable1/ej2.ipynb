```python
import numpy as np
import matplotlib.pyplot as plt

# Función para la restricción del conjunto X
def constraint1(x, y):
    return x**2 + y**2 <= 1

def constraint2(x, y):
    return 2*x - y <= 0

def constraint3(x, y):
    return y > 0.5

# Definición de la función de costo -log(y^2 - x^2)
def cost_function(x, y):
    return -np.log(y**2 - x**2)

# Crear una malla de puntos en el rango deseado
x_vals = np.linspace(0, 1, 400)
y_vals = np.linspace(0.5, 1.5, 400)
X, Y = np.meshgrid(x_vals, y_vals)

# Evaluar las restricciones y la función de costo en la malla de puntos
constraints = constraint1(X, Y) & constraint2(X, Y) & constraint3(X, Y)
cost_values = cost_function(X, Y)

# Crear la figura y los subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Graficar el conjunto X
ax1.contourf(X, Y, constraints, levels=[0, 1], colors=['lightgray', 'white'])
ax1.plot(x_vals, 2*x_vals, label=r'$2x - y \leq 0$')
ax1.plot(x_vals, np.sqrt(1 - x_vals**2), label=r'$x^2 + y^2 \leq 1$')
ax1.plot(x_vals, 0.5 * np.ones_like(x_vals), label=r'$y > 1/2$')
ax1.set_xlim(0, 1)
ax1.set_ylim(0.5, 1.5)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('Conjunto $X$')
ax1.legend()

# Graficar las curvas de nivel de la función de costo
contour = ax2.contour(X, Y, cost_values, levels=30, cmap='viridis')
ax2.clabel(contour, inline=1, fontsize=10)
ax2.set_xlim(0, 1)
ax2.set_ylim(0.5, 1.5)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Curvas de Nivel de $-\log(y^2 - x^2)$')

plt.tight_layout()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de la función de costo -log(y^2 - x^2)
def cost_function(x, y):
    return -np.log(y**2 - x**2)

# Crear una malla de puntos en el rango deseado
x_vals = np.linspace(0, 1, 400)
y_vals = np.linspace(0.5, 1.5, 400)
X, Y = np.meshgrid(x_vals, y_vals)

# Evaluar la función de costo en la malla de puntos
cost_values = cost_function(X, Y)

# Crear la figura
fig, ax = plt.subplots(figsize=(8, 6))

# Graficar la región factible
ax.fill_between(x_vals, 0.5, 2*x_vals, color='lightgray', label=r'$2x - y \leq 0$')
ax.fill_between(x_vals, np.sqrt(1 - x_vals**2), 1.5, color='lightgray', label=r'$x^2 + y^2 \leq 1$')
ax.fill_between(x_vals, 0.5, 1.5, where=(x_vals > 0.5), color='lightgray', label=r'$y > 1/2$')

# Graficar las curvas de nivel de la función de costo
contour = ax.contour(X, Y, cost_values, levels=20, cmap='viridis')
ax.clabel(contour, inline=1, fontsize=10)

# Etiquetas y título
ax.set_xlim(0, 1)
ax.set_ylim(0.5, 1.5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Función de Costo y Región Factible')
ax.legend()

plt.tight_layout()
plt.show()
```
